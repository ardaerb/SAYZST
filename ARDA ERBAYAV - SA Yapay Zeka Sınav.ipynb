### Soru 5
$f$ fonksiyonu Öklid Algoritması aracılığıyla $c$ değerini getirir.

$a$'yı 9, $b$'yi 22 olarak alırsak,
Dikdörtgene sığan 2 tane büyük $9 \times 9$'lik kare ve 2 tane küçük $4 \times 4$'lük kare ortaya çıkar ve arta kalan dikdörtgenimizin içine de 4 adet $1 \times 1$'lik kare girer. $f(a,b)=c=1$ olur.
Aynı zamanda 9 ve 22'nin EBOB'u 1'dir kanısına da varabiliriz.

$a=9$ ve $b=22$ için $f$'in yaptığı işlem basamaklarını
$$22=9(2)+4$$
$$9=4(2)+1$$
$$4=4(1)+0$$
olarak da gösterebiliriz.
